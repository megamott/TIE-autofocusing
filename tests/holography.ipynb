```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.restoration import unwrap_phase
```

```python
width, height = 256, 256
wavelength = 659.6e-9
k = 2 * np.pi / wavelength
px_size = 5.04e-6
```

```python
Y, X = np.mgrid[-height / 2:height / 2, -width / 2:width / 2]
Y, X = px_size * Y, px_size * X
radius_vector = np.sqrt(X ** 2 + Y ** 2)
radius_vector[radius_vector == 0] = 0.0000000000001

sn = X / radius_vector
field = np.exp(-1j * k * (np.sqrt(3) * (radius_vector ** 2 - 1) + (np.sqrt(8) * (3 * radius_vector ** 3 - 2 * radius_vector) * sn)))
field[radius_vector > 100] = 0

phase = np.angle(field)
unwrapped_phase = unwrap_phase(phase)
```

```python
plt.plot(unwrapped_phase[:, 128])
plt.show()
```

```python
plt.imshow(unwrapped_phase)
plt.colorbar()
plt.show()
```
