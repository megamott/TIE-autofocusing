```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from miscellaneous.optics import angular_spectrum_band_limited_1d
from skimage.restoration import unwrap_phase
```

```python
figure(figsize=(10, 8), dpi=100)
complex_field = np.load('../images/sin_0.5_intensity.npy')
phase = unwrap_phase(np.angle(complex_field))
plt.plot(np.angle(complex_field))
```

```python
start=0.000000  # 0 mkm
stop=0.0001  # 100 mkm
step=0.00005  # 50 mkm

px_size = 0.000005  # 5 mkm
wavelength = 0.000000555  # 555 nm

distances = np.arange(start, stop + step, step)
for distance in distances:
    wave_z = angular_spectrum_band_limited_1d(complex_field, distance, wavelength, px_size)
    filename = f'sin_z={distance * 1000:.3f}.npy'
    save_path = filename
    np.save(save_path, wave_z)
```
