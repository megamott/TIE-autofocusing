```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i_amplitude = 0.5
intensity = i_amplitude * np.ones([1024, 1024])
plt.imshow(intensity)
plt.colorbar()
plt.show()
```

```python
intensities_with_noise = {}
for standard_deviation in [0.01, 0.005, 0.0025, 0.001, 0.0001]:
    intensity = i_amplitude * np.ones([1024, 1024])

    mean = 0.0
    noise = np.random.normal(mean, standard_deviation, size=intensity.shape)
    intensity += noise
    intensities_with_noise[standard_deviation] = intensity
```

```python
plt.imshow(intensities_with_noise[0.005])
plt.colorbar()
plt.show()
```

  SNR     : $SNR=10*log_{10}[\frac{\sum_{0}^{n_x-1}\sum_{0}^{n_y-1} [r(x,y)]^2}{\sum_{0}^{n_x-1}\sum_{0}^{n_y-1} [r(x,y)-t(x,y)]^2]}$

```python
intensity_without_noise = i_amplitude * np.ones([1024, 1024])

for standard_deviation, intensity_with_noise in intensities_with_noise.items():
    SNR = 10 * np.log10(np.sum(intensity_with_noise**2)/np.sum((intensity_with_noise-intensity_without_noise)**2))
    print(f'Standard_deviation: {standard_deviation}. SNR: {SNR} dB.')
```
